```python
from transformers import BertTokenizer

f = open('./scene.text.txt', 'r', encoding='utf-8')
list = []
while True:
    line = f.readline()
    if not line: break
    scene = line.split('\t')[-1]
    list.append(scene.strip())

print(list)

f.close()

for i, sentence in enumerate(list):
    list[i] = sentence.replace('/', '')

tokenizer = BertTokenizer.from_pretrained("./", do_lower_case=False)
tokenizer(
    text=sentence,
    add_special_tokens=True,
    max_length=45,
    return_tensors='pt'
)
tokenized_sentences = [tokenizer.tokenize(sentence) for sentence in list]
```
