```python
f = open('./scene.txt', 'r')
list = []
while True:
    line = f.readline()
    if not line: break
    print(line.split('\t'))
    scene = eval(line.split('\t')[-1])
    list.append(scene)

mem = [-1,0,0,0]

for aScene in list:
    mem[len(aScene)] = mem[len(aScene)] + 1

print(mem)
f.close()
```

```python
f = open('./scene.text.txt', 'r', encoding='utf-8')
list = []
while True:
    line = f.readline()
    if not line: break
    # print(line.split('\t'))
    scene = line.split('\t')[-1]
    list.append(scene.strip())

print(list)

f.close()
```

```python
for i, sentence in enumerate(list):
    list[i] = sentence.replace('/', '')
```

```python
print(list)
```

```python
from transformers import BertTokenizer
tokenizer_bert = BertTokenizer.from_pretrained("./", do_lower_case=False)
```

```python
tokenized_sentences = [tokenizer_bert.tokenize(sentence) for sentence in list]
```

```python
for i in tokenized_sentences:
    print(i)
```

```python
size = {}
for i in tokenized_sentences:
    if size.get(len(i), None) is None:
        size[len(i)] = 1
    else:
        size[len(i)] += 1

print(size)
print(max(size.keys()))
```

```python
for i, sentence in enumerate(tokenized_sentences):
    for token in sentence:
        if token == '[UNK]':
            print('shit', i, sentence, token)
```
